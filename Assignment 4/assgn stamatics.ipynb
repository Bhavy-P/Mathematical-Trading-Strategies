```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

# AAPL

```python
data=yf.download("AAPL", start='2013-07-05', end='2023-07-05')
```

```python
data['SMA_50']=data['Close'].rolling(window=50).mean()
data['SMA_200']=data['Close'].rolling(window=200).mean()
```

## Bull Flag 

```python
bf= (data['Close'].rolling(window=10).max() == data['Close'])
data['bull_flag']=bf.astype(int)
```

```python
plt.figure(figsize=(15, 10))
plt.plot(data.index, data['Close'], label='( CLOSE )')
plt.scatter(data[data['bull_flag']==1].index, data[data['bull_flag']==1]['Close'], marker='X', color='red')
plt.xlabel('time')
plt.ylabel("Close")
plt.xticks(rotation=45)
plt.show()
```

## Inverse Cup & Handle

```python
inverse_cup_handle = (data['Low'].rolling(window=30).min() == data['Low']) & (data['SMA_50'] < data['SMA_200'])
data['inv_cup_and_handle'] = inverse_cup_handle.astype(int)
```

```python
plt.figure(figsize=(15, 10))
plt.plot(data.index, data['Close'], label='( CLOSE )')
plt.scatter(data[data['inv_cup_and_handle']==1].index, data[data['inv_cup_and_handle']==1]['Close'], marker='X', color='red')
plt.xlabel('time')
plt.ylabel("Close")
plt.xticks(rotation=45)
plt.show()
```

## Descending Triangle

```python
descending_triangle = (data['High'].rolling(window=30).max() == data['High']) & (data['Low'].rolling(window=30).min().shift(1) == data['Low'])
data['desc_tri'] = descending_triangle.astype(int)
```

```python
plt.figure(figsize=(15, 10))
plt.plot(data.index, data['Close'], label='( CLOSE )')
plt.scatter(data[data['desc_tri']==1].index, data[data['desc_tri']==1]['Close'], marker='X', color='red')
plt.xlabel('time')
plt.ylabel("Close")
plt.xticks(rotation=45)
plt.show()
```

## Rounding Bottom

```python
rounding_bottom = (data['Low'].rolling(window=30).min() == data['Low']) & (data['Close'].rolling(window=10).max() == data['Close'])
data['round_bot'] = rounding_bottom.astype(int)
```

```python
plt.figure(figsize=(15, 10))
plt.plot(data.index, data['Close'], label='( CLOSE )')
plt.scatter(data[data['round_bot']==1].index, data[data['round_bot']==1]['Close'], marker='X', color='red')
plt.xlabel('time')
plt.ylabel("Close")
plt.xticks(rotation=45)
plt.show()
```

## Inverse Head & Shoulders

```python
inverse_head_shoulder = (data['Low'].rolling(window=30).min() == data['Low']) & (data['Close'].rolling(window=10).max() == data['Close'])
data['inv_head_shoulder'] = inverse_head_shoulder.astype(int)
```

```python
plt.figure(figsize=(15, 10))
plt.plot(data.index, data['Close'], label='( CLOSE )')
plt.scatter(data[data['inv_head_shoulder']==1].index, data[data['inv_head_shoulder']==1]['Close'], marker='X', color='red')
plt.xlabel('time')
plt.ylabel("Close")
plt.xticks(rotation=45)
plt.show()
```
